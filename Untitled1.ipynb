```python
# YOLOv8 설치 및 기본 사용법
# Google Colab 환경에서 실행하는 코드입니다.

# 1. 필요한 패키지 설치
!pip install ultralytics

# 2. 필요한 라이브러리 임포트
import cv2
import numpy as np
from ultralytics import YOLO
from google.colab import files
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# 3. YOLOv8 모델 로드 (기본 모델 - YOLOv8n)
model = YOLO('yolov8n.pt')  # 'n'은 nano 버전으로 가장 가벼운 모델입니다.
# 다른 크기의 모델: 'yolov8s.pt' (small), 'yolov8m.pt' (medium), 'yolov8l.pt' (large), 'yolov8x.pt' (x-large)
```

```python
# 업로드된 이미지에서 객체 감지 실행
results = model('1.1.jpg')

# 결과 시각화
for r in results:
    # 감지 결과가 포함된 이미지 출력
    im_array = r.plot()
    cv2_imshow(im_array)

    # 감지된 객체 정보 출력
    print("\n감지된 객체:")
    for box in r.boxes:
        cls = int(box.cls[0])
        conf = float(box.conf[0])
        print(f"- {model.names[cls]}: {conf:.2f} 신뢰도")
```
